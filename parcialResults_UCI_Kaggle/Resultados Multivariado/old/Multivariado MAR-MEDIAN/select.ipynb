```python
df = pd.read_csv("MAR-MEDIAN.csv")

df.head()
```

```python
df = df.drop(columns='Unnamed: 0')

cols = df.columns

cols
```

```python
df.head()
```

```python
for i in cols:
    if (i != 'Dataset') and (i != 'Missing Rate') :
        df[i] = df[i].str.split().str[0]


df.head()
```

```python
df.info()
```

```python
for i in cols:
    if (i != 'Dataset'):
        df[i] = df[i].astype(float)


df.head()
```

```python
df.info()
```

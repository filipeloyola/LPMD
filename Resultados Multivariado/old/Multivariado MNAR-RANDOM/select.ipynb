```python
import pandas as pd
```

```python
df = pd.read_csv("Resultados Multivariado MNAR-MBOUV_filtrado.csv")

df = df.drop(columns='Unnamed: 0')

cols = df.columns

cols
```

```python
for i in cols:
    if (i != 'Dataset') and (i != 'Missing Rate') :
        df[i] = df[i].str.split().str[0]


df
```

```python
df.info()
```

```python
for i in cols:
    if (i != 'Dataset'):
        df[i] = df[i].astype(float)


df
```

```python
df.to_csv("Resultados_algoritmos.csv")
```
